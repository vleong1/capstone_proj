## Make Smaller File

```python
# pickle whole file, including essay columns? -- wayy too large still
# large file not saved

cupid.to_pickle('data/cupid.pkl')
```

```python
cupid_check = pd.read_pickle('data/cupid.pkl')
```

```python
cupid_check.info()
```

```python
# drop essay columns
cupid_subset = cupid.iloc[:,:21]
```

```python
cupid_subset.info()
```

```python
# change object type to category
# referenced https://git.generalassemb.ly/DSI-322/7.01-big-data/blob/master/big-data-memory-post-class.ipynb

cat_cols = cupid_subset.drop(columns = 'height').columns
cupid_subset[cat_cols] = cupid_subset[cat_cols].astype('category')
```

```python
cupid_subset.info()
```

```python
cupid_subset['status'].value_counts()
```

```python
# drop columns discriminating columns (i.e. ethnicity, height), unnecessary columns ('last_online', 'speaks'), and odd columns that had weird values ('income', 'job')

cupid_subset.drop(columns = ['height', 'ethnicity', 'job', 'last_online', 'speaks', 'income'],inplace = True)
```

```python
cupid_subset.info()
```

```python
cupid_subset.to_pickle('data/subset_cupid.pkl')
```
